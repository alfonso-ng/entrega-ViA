# Actividad

Detector de movimiento en una región de interés (ROI) marcada a mano, que
guarda un vídeo de cada evento, clasifica los objetos que aparecen y envía
una notificación por correo solo si pertenecen a una categoría de interés.
Si en el evento aparece una persona, se anonimiza su cara.

## Cómo funciona

MOG2 sobre la ROI marcada con el ratón → si hay suficiente movimiento durante varios frames seguidos, se arranca la grabación. Se mantiene un preroll de 1 segundo para no perder el inicio del evento. El evento acaba cuando pasan unos segundos sin movimiento.

Mientras graba, cada ciertos frames se pasa el recorte de la ROI por EfficientDet-Lite (el mismo de `mp_object/objectrecog.py`) para ver qué categorías COCO aparecen. Si alguna coincide con `--target`, se manda un correo con una captura. Si aparece una persona, se le tapa la cara antes de enviar.

Todo queda anotado en `actividad.csv`. Script: **`actividad.py`**, con **`correo.py`** para el envío.

### Uso

```
./actividad.py --dev=0 --target=bird,cat
```

Arrastrar el ratón para marcar la ROI. `c` para borrarla. `m` para ver la máscara. Mientras graba aparece un círculo rojo. El correo se configura en `correo.env` (copiar de `correo.env.example`).
